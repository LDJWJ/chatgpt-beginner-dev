### A. if문의 기본 문법


````python
## if문
x = 5

# if (  ):
# ___실행문1
if x > 0:
  print("x는 양수입니다.")
````

````python
### if-else문
x = 5

if x>0:
  print(f"{x}은 양수입니다.")
else:
  print(f"{x}은 음수입니다.")
````

````python
## if-elif-else문
score = 85

if score>=90:
  print("A학점, 우수합니다!")
elif score >=80:
  print("B학점. 잘했어요!")
elif score >=70:
  print("C학점. 노력해 주세요.")
else:
  print("D학점. 많이 노력해 주세요.")
````

````python
## 중첩 if문
num = 15
if num>=10:
  if num % 2 ==0:
    print("10보다 크고 짝수")
  else:
    print("10보다 크고, 홀수")

````

````python
## 조건식이 여러 개인 if
age = 20
membership = True
if age >= 18 and membership:   # 조건식1 or 조건식2
    print("당신은 특별 할인 대상입니다.")
````

### C. for문의 기본 문법

# 기본 구조
```
for 변수 in 반복가능한객체:
  # 실행할 코드 블록
```

````python
fruits = ['apple', 'banana', 'cherry']

for one in fruits:
  print(one)
````

````python
### 기본 구조 2
# range()
for i in range(1,11,1):
  print(i)
````

### D. for문의 여러가지 예제

````python
## 리스트 순회
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
````

````python
# 문자열 순회
for char in "hello":
  print(char)
````

````python
## 딕셔너리 순회
info = {"name": "John", "age": 30}
for key in info:
    print(key, info[key])
````

````python
# enumerate() 함수 사용
fruits = ["apple", "banana", "cherry"]
for idx, fruit in enumerate(fruits):
  print(idx, fruit)
````

````python
## 중첩 for문
for i in range(3):
  print(i)
  for j in range(2):
    print(f"i={i}, j={j}")
````

````python
## 리스트 컴프리헨션 - for문을 사용한 간단한 리스트 생성
## 구조 : [expression for item in iterable]
squares = [x*2 for x in range(10)]
print(squares)
````

````python
#
squares = [x*2+1 for x in range(10)]
print(squares)
````

````python
# 과일 찾기
my_list = ["사과", "바나나", "참외", "포도", "딸기"]

find_fruit = "참외"

for fruit in my_list:
    if fruit == find_fruit:
        print(f"{find_fruit}는 리스트에 있습니다.")
        break
````
