### IPython Audio 클래스

* IPython.display.Audio 클래스는 IPython (주피터 노트북 포함) 환경에서 오디오 파일을 로드하고 재생하는 데 사용되는 도구입니다.

* Audio 클래스의 주요 기능
 - 오디오 파일 로드 : 다양한 오디오 파일 형식(예:mp3, wav, ogg등)을 로드할 수 있습니다.
 - 오디오 재생 : 로드된 오디오 파일을 주피터 노트북 내에서 재생할 수 있습니다.
 - 재생 제어 : 재생/일시정지, 시간 이동 등의 기본적인 재생 제어 기능을 제공합니다.
 - Audio 클래스는 데이터 과학, 음성 처리, 음악 이론 교육 등 다양한 분야에서 유용하게 사용합니다.

### 오디오 플레이어를 만들어보기

````python
from IPython.display import Audio
import os

files = os.listdir()
mp3_files = [f for f in files if f.endswith(".mp3")]

print("아래 음악 리스트에서 재생하고 싶은 음악을 선택해 주세요.")

cnt = 0
for one in mp3_files:
  cnt += 1
  print(cnt, ":", one)

sel = int( input("선택할 오디오 번호"))

# 오디오 파일 지정과 재생
music_file = mp3_files[sel-1]

Audio(music_file)
````
