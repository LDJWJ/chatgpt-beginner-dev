### (1) matplotlib과 seaborn은 무엇일까요?

### Matplotlib
 - Matplotlib은 Python의 가장 인기 있는 데이터 시각화 라이브러리 중 하나입니다.
 - 2D 및 3D 그래프를 쉽게 생성할 수 있도록 설계되었으며, 다양한 운영 체제와 그래픽 백엔드에서 사용할 수 있습니다.
 - Matplotlib을 사용하면 과학적인 계산을 위한 그래프, 차트, 플롯 등을 높은 수준의 커스터마이징으로 그릴 수 있습니다.
 - 이 라이브러리는 John D. Hunter에 의해 처음 개발되었고, MATLAB과 유사한 인터페이스를 제공하여 많은 사용자들이 쉽게 접근할 수 있도록 하였습니다.

### Matplotlib의 장단점

 ### [장점]
 - 유연성 : 다양한 그래프 유형 및 기능 제공(선 그래프, 막대 그래프, 산점도, 히스토그램, 파이 차트 등)
 - 확장성 :  색상, 스타일, 레이아웃 등을 세밀하게 조정할 수 있음
 - 다양한 출력 포맷: PNG, PDF, SVG, EPS 등 다양한 포맷으로 그래프를 저장.

 ### [단점]
  - 상대적인 난이도 : 초보자에게는 그래프를 그리기 위한 난이도가 다소 높을 수 있다.
  - 코드의 작성량 : 원하는 시각화를 위해 직접 코드 작성이 필요.


### Seaborn

 - Seaborn은 Matplotlib을 기반으로 하는 또 다른 Python 시각화 라이브러리로, 통계적 그래프를 생성하는 데 중점을 둡니다.
 - Seaborn은 Matplotlib의 기능을 확장하고, 데이터 구조와 함께 작동하도록 설계되어 더 아름답고 복잡한 시각화를 쉽게 생성할 수 있습니다.
 - Seaborn은 데이터의 패턴을 이해하고, 여러 변수 간의 관계를 시각화하며, 분포를 시각화하는 데 특히 유용합니다.

### Seaborn의 장단점

### [장점]
  - 간편성: 간단한 코드로 매력적인 그래프 제작 가능
  - 효율성: 데이터 시각화 작업 시간 단축
  - 미적 요소: 기본적으로 깔끔하고 전문적인 시각화 제공
  - 호환성 : Pandas DataFrame과 같은 데이터 구조와의 긴밀한 통합으로 복잡한 데이터 세트를 쉽게 처리할 수 있음.
  - 확장성 : 회귀선이나 상관관계 등을 보여주는 복잡하 시각화가 간단하게 가능.

### [단점]
   - 유연성 제한: Matplotlib에 비해 기능 및 그래프 유형 선택 폭이 다소 좁음
   - 커스터마이징 제한: Matplotlib에 비해 세밀한 조정 기능이 다소 제한적

### 어떤 경우에 어떤 그래프를 사용할까?

### Matplotlib
 - 고도로 맞춤화된 그래프 제작
 - 특정 기능이나 그래프 유형이 필요한 경우
 - 데이터 시각화에 대한 깊은 이해를 가진 사용자

### Seaborn
 - 기본적인 그래프를 빠르고 간편하게 제작
 - 깔끔하고 전문적인 시각화를 원하는 경우
 - 데이터 분석 프로젝트에 시각화 기능 추가

Matplotlib과 Seaborn의 관계는 서로 보완적입니다. Matplotlib은 그래프의 기본적인 구조와 스타일링을 제공하는 반면, Seaborn은 복잡한 데이터 구조를 더 쉽고 아름답게 시각화할 수 있는 고급 인터페이스를 제공합니다. 데이터 시각화에 있어서 이 두 라이브러리를 함께 사용하면 효과적인 데이터 분석과 통찰력 있는 시각화를 생성할 수 있습니다.

### (2) 기본 구조 이해 및 기본 그래프 그리기
  - 기본 구조 이해 - Figure, Axes, Axis
    - Figure : 전체 그래프의 컨테이너 역할. Figure는 그래프의 창이나 페이지를 의미하며, 여러 그래픽 요소들을 포함할 수 있는 화폭과 같다.
    - Axes : 이것은 실제로 데이터가 표시되는 영역이다. 각각의 Axes에는 하나 이상의 'Axis'가 포함되어 있다. 이것이 실제로 데이터 플롯을 나타내는 부분이다. Axis는 우리가 '그래프'라 부르고, 좌표축이 있는 플롯 공간을 가르킨다.
    - Axis : 이것들은 수평축(x-axis)과 수직축(y-axis)을 나타내는 객체이다. 축에는 눈금(tick)과 눈금 라벨(tick label)이 포함되어 있다. 데이터의 스케일을 설정하는 데 사용.

```python
# ==========================================
import matplotlib.pyplot as plt

# Figure: 전체 그래프 영역
fig = plt.figure()

# Axes: 그래프를 담는 공간 (하나 이상 포함 가능)
ax = fig.add_subplot(111)

# Axis: x축, y축 설정
ax.set_xlabel("x축")
ax.set_ylabel("y축")

# 그래프 표시
plt.show()
```

### (2) 기본 구조 이해 및 기본 그래프 그리기
 - 기본그래프 그리기 - 선 그래프, 막대 그래프, 히스토그램, 산점도

```python
# ==========================================
## 선 그래프
import matplotlib.pyplot as plt

# 데이터 생성
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# 선 그래프 그리기
plt.plot(x, y, 'o')

# 그래프 표시
plt.show()
```

```python
# ============================================
## 막대 그래프
import matplotlib.pyplot as plt

# 데이터 생성
x = ["A", "B", "C", "D", "E"]
y = [10, 20, 30, 40, 50]

# 막대 그래프 그리기
plt.bar(x, y)

# 그래프 표시
plt.show()
```

```python
# ============================================
## 막대 그래프
categories = ['Category A', 'Category B', 'Category C']
values = [1, 2, 3]

plt.bar(categories, values)
plt.title('Bar Graph')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# ============================================
## 히스토그램
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
data = np.random.randn(100)

# 히스토그램 그리기
plt.hist(data)

# 그래프 표시
plt.show()
```

```python
# ============================================
### 산점도
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.random.rand(100)
y = np.random.rand(100)

# 산점도 그리기
plt.scatter(x, y)

# 그래프 표시
plt.show()
```

### (3) 그래프의 기본 요소 추가
 - 축, 제목, 레이블, 범례를 추가하기

```python
# ==========================================
### 축, 제목
import matplotlib.pyplot as plt

# 데이터 생성
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# 선 그래프 그리기
plt.plot(x,y)

# 축 설정
plt.xlabel("x축")
plt.ylabel("y축")

# 제목 설정
plt.title("제목")

# 그래프 표시
plt.show()
```

```python
# ==========================================
### 범례1
import matplotlib.pyplot as plt

# 데이터 생성
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [4, 8, 12, 16, 20]

# 선 그래프 그리기
plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2")

# 범례 표시
plt.legend()

# 그래프 표시
plt.show()
```

```python
# ==========================================
### 범례2
import numpy as np
x = np.linspace(0, 10, 100)
x

```

```python
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, '-b', label='Sine')
plt.plot(x, y2, '-r', label='Cosine')
plt.legend(loc='upper right')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Graph with Legends')
plt.show()
```
