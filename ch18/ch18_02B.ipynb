## Seaborn

 - Seaborn은 Matplotlib을 기반으로 하는 또 다른 Python 시각화 라이브러리로, 통계적 그래프를 생성하는 데 중점을 둡니다.
 - Seaborn은 Matplotlib의 기능을 확장하고, 데이터 구조와 함께 작동하도록 설계되어 더 아름답고 복잡한 시각화를 쉽게 생성할 수 있습니다.
 - Seaborn은 데이터의 패턴을 이해하고, 여러 변수 간의 관계를 시각화하며, 분포를 시각화하는 데 특히 유용합니다.

### Seaborn의 장단점
### [장점]
  - 간편성: 간단한 코드로 매력적인 그래프 제작 가능
  - 효율성: 데이터 시각화 작업 시간 단축
  - 미적 요소: 기본적으로 깔끔하고 전문적인 시각화 제공
  - 호환성 : Pandas DataFrame과 같은 데이터 구조와의 긴밀한 통합으로 복잡한 데이터 세트를 쉽게 처리할 수 있음.
  - 확장성 : 회귀선이나 상관관계 등을 보여주는 복잡하 시각화가 간단하게 가능.

### (1) 수치형(히스토그램), 범주형(막대 그래프, Boxplot) 그리기

```python
# ==========================================
### [라이브러리 불러오기]
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 한글 폰트 설정
sns.set_theme(font="NanumGothic")
```

```python
###  [수치형 데이터 시각화] - 히스토그램 그리기
# 데이터 생성
# 평균이 0이고, 표준이 1인 표준 정규 분포를 따르는 무작위 샘플 100개 생성
data1 = np.random.randn(100)

# 히스토그램 그리기
sns.histplot(data = data1)
plt.show()
```

```python
### [범주형 데이터 시각화] - 막대그래프 그리기
# 데이터 생성
categories = ["A", "B", "C", "D", "E"]
values = [10, 20, 30, 40, 50]

# 막대 그래프 그리기
sns.barplot(x=categories, y=values)
plt.show()
```

```python
# 데이터 생성
df = sns.load_dataset("iris")
df.columns
```

```python
### Boxplot 시각화
# Boxplot 그리기
sns.boxplot(x="species" , y="sepal_length" , data=df )
```

```python
### 영화 IMDB데이터
# 라이브러리 불러오기
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# IMDB 데이터셋 가져오기
imdb_url = "https://raw.githubusercontent.com/thechaudharysab/imdb-data-pandas-visualization/master/data/imdb_1000.csv"

# 데이터 가져오기
movies = pd.read_csv(imdb_url)  # 또는 다른 적절한 인코딩 사용
movies

# 특수 문자 제거
movies['actors_list'] = movies['actors_list'].str.replace(r"[\[u\'\]]", '')  # 특수 문자 제거

# 컬럼명 확인
print(movies.columns)

# 데이터셋 미리보기
movies.head()
```

```python
# 상영시간 분포 확인하기
sns.histplot(x='duration', data=movies)
plt.show()
```

```python
# 장르별 영화 편수 확인
sns.countplot(x='genre', data=movies)
plt.xticks(rotation=45)
plt.show()
```

```python
# 장르별 평균 평점 계산
genre_ratings = movies.groupby("genre")["star_rating"].mean()
genre_ratings
```

```python
# 장르별 평균 평점 계산
genre_ratings = movies.groupby("genre")["star_rating"].mean()

# 장르별 상관관계 히트맵 생성
sns.heatmap(genre_ratings.to_frame(), annot=True)
plt.xlabel("Genre")
plt.ylabel("Average Rating")
plt.title("Genre-wise Average Rating")
plt.show()
```
