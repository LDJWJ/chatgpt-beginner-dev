 * 만약 Google colab 등에서 작업을 수행할 경우, 캐글에서 데이터를 다운로드 후, 데이터를 구글 코랩에 업로드 후, 데이터 불러오기 경로를 수정해야 한다.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
# 데이터 불러오기
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
submission = pd.read_csv("/kaggle/input/titanic/gender_submission.csv")
```

```python
# 간단한 데이터 전처리
# 여기서는 'Sex' 컬럼을 숫자로 변환하고, 'Age'와 'Fare'의 결측값을 중앙값으로 대체
train_data['Sex'] = train_data['Sex'].map({'male': 0, 'female': 1})
test_data['Sex'] = test_data['Sex'].map({'male': 0, 'female': 1})

train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
test_data['Age'].fillna(test_data['Age'].median(), inplace=True)

train_data['Fare'].fillna(train_data['Fare'].median(), inplace=True)
test_data['Fare'].fillna(test_data['Fare'].median(), inplace=True)
```

```python
# 사용할 특성 선택
features = ["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare"]

X = train_data[features]
y = train_data["Survived"]
X_test = test_data[features]

# 데이터 스케일링
scaler = StandardScaler()
X = scaler.fit_transform(X)
X_test = scaler.transform(X_test)

# 로지스틱 회귀 모델 훈련
model = LogisticRegression()
model.fit(X, y)
```

```python
# 테스트 데이터셋에 대한 생존 예측
predictions = model.predict(X_test)



# 제출 파일 준비
submission['Survived'] = predictions
submission.to_csv("my_submission.csv", index=False)
print("Your submission was successfully saved!")
```

### 제출 결과
 * 정확도 : 0.75358 (75.35%)

### 추가 ChatGPT AI 생성 코드

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler


# 데이터 불러오기
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
submission = pd.read_csv("/kaggle/input/titanic/gender_submission.csv")
```

```python
# 간단한 데이터 전처리
# 여기서는 'Sex' 컬럼을 숫자로 변환하고, 'Age'와 'Fare'의 결측값을 중앙값으로 대체합니다.
train_data['Sex'] = train_data['Sex'].map({'male': 0, 'female': 1})
test_data['Sex'] = test_data['Sex'].map({'male': 0, 'female': 1})

train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
test_data['Age'].fillna(test_data['Age'].median(), inplace=True)

train_data['Fare'].fillna(train_data['Fare'].median(), inplace=True)
test_data['Fare'].fillna(test_data['Fare'].median(), inplace=True)



# 사용할 특성 선택
features = ["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare"]

X_train = train_data[features]
y_train = train_data["Survived"]
X_test = test_data[features]

# 데이터 스케일링
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)



# 랜덤 포레스트 모델 훈련
model = RandomForestClassifier(n_estimators=100, 
                               max_depth=5, 
                               random_state=1)
model.fit(X_train, y_train)



# 테스트 데이터셋에 대한 생존 예측
predictions = model.predict(X_test)


# 제출 파일 준비
submission['Survived'] = predictions
submission.to_csv("my_rf_submission.csv", index=False)

print("Your submission was successfully saved!")
```
