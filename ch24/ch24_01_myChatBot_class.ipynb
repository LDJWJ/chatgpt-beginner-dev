```python
# ëª¨ë¸ - GPT 3.5 Turbo ì„ íƒ
# OpenAI GPT ëª¨ë¸ ì¤‘ gpt-3.5-turboë¥¼ ì„ íƒí•©ë‹ˆë‹¤.
# ì´ ëª¨ë¸ì€ OpenAIì˜ ìµœì‹  ì–¸ì–´ ëª¨ë¸ ì¤‘ í•˜ë‚˜ë¡œ, ë†’ì€ ì •í™•ë„ì˜ ì‘ë‹µì„ ìƒì„±í•˜ëŠ” ë° ì í•©í•©ë‹ˆë‹¤.
model = "gpt-3.5-turbo"

# ì§ˆë¬¸ ì‘ì„±í•˜ê¸°
# ì‚¬ìš©ìë¡œë¶€í„° ë°›ì•„ë‚¼ ì§ˆë¬¸ì„ query ë³€ìˆ˜ì— í• ë‹¹
query = "ìš´ë™ ì™•ì´ˆë³´ì¸ë° 5ì¼ê°„ì˜ ìš´ë™ ìŠ¤ì¼€ì¤„ ë‹µë³€ì„ ë¶€íƒí•´. í•œê¸€ë¡œ ë‹µë³€"
```

```python
# ChatGPT API í˜¸ì¶œí•˜ê¸°
# API í˜¸ì¶œì„ í†µí•´ modelê³¼ msg1 ë©”ì‹œì§€ ë¦¬ìŠ¤íŠ¸ë¥¼ ì‚¬ìš©í•˜ì—¬ ëŒ€í™”ë¥¼ ìƒì„±
# client.chat.completions.create()ëŠ” OpenAI APIì˜ completions ì—”ë“œí¬ì¸íŠ¸ë¥¼ ì‚¬ìš©í•´ ë‹µë³€ì„ ìƒì„±í•˜ëŠ” ë©”ì„œë“œ
response = client.chat.completions.create(
    model=model,
    messages=msg1
)
```
