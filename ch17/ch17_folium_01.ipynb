````python
### 1-1 기본 지도 생성
# 라이브러리 불러오기
import folium

# 지도 객체 생성
map = folium.Map(location=[37.5665, 126.9780], zoom_start=13)

# 지도 출력
map
````

````python
### 1-2 마커 추가

import folium

# 기본 지도 생성: 지도의 중심을 서울 시청(위도: 37.5665, 경도: 126.9780)으로 설정하고, 줌 레벨은 13으로 설정
map_seoul = folium.Map(location=[37.5665, 126.9780], zoom_start=13)

# 지도에 마커 추가: 서울 시청 위치에 마커 추가, 팝업 텍스트는 '서울 시청'
folium.Marker([37.5665, 126.9780], popup='서울 시청').add_to(map_seoul)

# 지도 표시
map_seoul
````

### (2) 시각화 실습 및 데이터 가져오기

````python
### 2-1 경로 그리기
import folium

# 지도 객체 생성
map = folium.Map(location=[37.5665, 126.9780], zoom_start=13)

# 경로 지정
locations = [[37.5665, 126.9780], [37.5665, 126.9980], [37.5565, 126.9980]]

# 경로 그리기
folium.PolyLine(locations).add_to(map)


map
````

````python
### 2-2 지도 저장
import folium

# 지도 객체 생성
m = folium.Map(location=[37.5665, 126.9780], zoom_start=13)

# 경로 지정
locations = [[37.5665, 126.9780], [37.5665, 126.9980], [37.5565, 126.9980]]

# 경로 그리기
folium.PolyLine(locations).add_to(m)


# 지도 출력
m.save("basic_map.html")

m
````

````python
### 2-3. 여러 개의 마커 추가
import folium

# 지도 객체 생성
m = folium.Map(location=[37.366535, 128.977969], zoom_start=7)

# 여러 개의 마커 객체 생성 및 지도에 추가
markers = [
    folium.Marker(location=[37.566535, 126.977969], popup="서울"),
    folium.Marker(location=[37.527256, 127.041902], popup="경기"),
    folium.Marker(location=[35.179554, 129.075941], popup="부산"),
]

for marker in markers:
  m.add_child(marker)

# 지도 출력
m.save("multiple_markers.html")

m
````

````python
### 공공 데이터 가져와 데이터 탐색
import pandas as pd

# 한국어 인코딩을 고려하여 데이터 다시 로드
file_path = "data.csv"
data_kr = pd.read_csv(file_path, encoding='euc-kr')
data_kr
````

# 위도 경도
```
위도 = Latitude
 위도는 지구 적도를 기준으로 북쪽 또는 남쪽으로 떨어진 각도를 나타내며,
경도 = Longitude
 경도는 본초자오선(그리니치 천문대를 지나는 자오선)을 기준으로 동쪽 또는 서쪽으로 떨어진 각도를 나타냅니다.
```

````python
### 라이브러리 불러오기
import seaborn as sns
import matplotlib.pyplot as plt

# 위도, 경도 정보 시각화
plt.figure(figsize=(10,8))
ax = sns.scatterplot(x='경도', y='위도', data=data_kr)
plt.xlabel('Longitude')  # 경도
plt.ylabel('Latitude')   # 위도
````
