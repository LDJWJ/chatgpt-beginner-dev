### 기본 자료형의 이해 및 실습

```python
## 문자열
greeting = 'Hello, World!'
print(greeting)
```

```python
s1 = "Hello, world!"
s2 = "파이썬은 재미있다."

print(s1, s2)
print(s1[0])  # H
print(s1[-1])
print(len(s2))
```

```python
# 정수형(integer), 실수형
age = 30
score = 50.10

print(age, score)
print(type(age), type(score) )
```

```python
# 자료형의 형변환
# int( ), str( ), float( ), list( ), dict( )

# 정수 -> 문자열
str_num = str(10)
print(str_num)
print(type(str_num) )
```

```python
### 문자열 -> 정수
str_num = '123'
int_num = int(str_num)
print(int_num) # 123
print(type(int_num) )
```

```python
age = int( input("당신의 나이는?") )
print(type(age))
print(age + 10)
```

```python
# 문자열 -> 실수
str_num = '123'
float_num = float(str_num)
print(float_num)
```

### 리스트, 딕셔너리 자료형 이해 및 실습

```python
fruits = [ ]
dict1 = { }
print(type(fruits), type(dict1))
```

```python
fruits = ['apple', 'banana', 'cherry']
print(fruits)
print(fruits[0])
print(fruits[0:2])
print(fruits[0:])
print(fruits[:2])
print(len(fruits))
```

```python
for one in fruits:
  print(one)
```

```python
person = {"name":"John", "age":30 }
print(person)
print(person['name'])
print(person['age'])

person['email'] = "toto@naver.com"
person
```

```python
## 리스트에 값 추가
fruits.append("orange")  # remove, extend
fruits
```
