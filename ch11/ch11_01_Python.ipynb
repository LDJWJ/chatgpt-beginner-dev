### 변수 이해 및 실습

```python
a = 10
print(a)
```

```python
a1 = "Hello World"
print(a1)
```

```python
a1 = 10 + 30
print(a1)
```

```python
a1 = "Hello"
a2 = "World"
a3 = a1 + a2
a4 = str(10) + "world"
print(a3)
print(a4)
```

### input() 함수 이해 및 실습

```python
a = input("당신이 좋아하는 과일은 무엇입니까?")
print(a)
```

```python
age = int( input("당신의 나이는?") )
print("당신의 10년후의 나이는 ", age+10, "살입니다.")
```

```python
age = int( input("당신의 나이는?") )
print(f"당신의 10년후의 나이는 {age+10}살입니다.")
```
