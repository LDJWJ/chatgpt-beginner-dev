```python
## 인증
from openai import OpenAI
from getpass import getpass
import os

# openai_key = "[나의 API KEY]"
openai_key = getpass("오픈AI에서 발급받은 인증키를 입력해 주세요.")
```

```python
# 모델 - gpt-3.5-turbo-0125 선택
os.environ["API_KEY"] = openai_key
client = OpenAI(api_key  = os.environ.get("API_KEY") )

# 모델 - GPT 3.5 Turbo 선택
model = "gpt-3.5-turbo-0125"
```

```python
# 질문 작성하기
# 사용자로부터 받아낼 질문을 query 변수에 할당
query = "운동 왕초보인데 5일간의 운동 스케줄 답변을 부탁해. 한글로 답변"


# 첫 번째 메시지는 "system" 역할로, AI에게 대화 스타일을 설정하는 데 사용.
# 두 번째 메시지는 "user" 역할로, 사용자의 실제 질문인 query를 전달.
msg1 =[
    {
      "role": "system",
      "content": "친절하게 답변 부탁해."
    },
    {
      "role": "user",
      "content": query
    }
]

# ChatGPT API 호출하기
# API 호출을 통해 model과 msg1 메시지 리스트를 사용하여 대화를 생성
# client.chat.completions.create()는 OpenAI API의 completions 엔드포인트를 사용해 답변을 생성하는 메서드
response = client.chat.completions.create(
    model=model,
    messages=msg1
)

# response 객체로부터 choices의 첫 번째 항목에서 message의 content를 추출하여 answer 변수에 저장
answer = response.choices[0].message.content
print( answer )
```

### 최종 AI 비서 완성해 보기

```python
from openai import OpenAI
from getpass import getpass
import os

# openai_key = "[나의 API KEY]"
openai_key = getpass("오픈AI에서 발급받은 인증키를 입력해 주세요.")
```

```python
os.environ["API_KEY"] = openai_key
client = OpenAI(api_key  = os.environ.get("API_KEY") )
model = "gpt-3.5-turbo-0125"
```

```python
## 사용자로부터 파라미터 값 입력
def get_user_settings():
  creativity = float(input("창의성 수준 선택 : (0.1 ~ 1.0)"))
  response_length = int(input("응답 길이 선택 : (50 ~ 700)"))
  stop_pharse = input("대화를 종료할 구문 입력 (예:'종료')")

  return creativity, response_length, stop_pharse

creativity, response_length, stop_phrase = get_user_settings()
```

```python
### 연속 대화 함수

def chat(creativity, response_length, stop_phrase):
  chat_history = []  # 대화 기록 저장
  print("대화를 시작하세요! (대화 종료 구문을 입력하면 대화가 종료됩니다.)")

  while True:
    user_input = input("사용자 : ")
    if user_input == stop_phrase:
      print("대화를 종료합니다.")
      break
    chat_history.append({"role":"user", "content":user_input})

    res = client.chat.completions.create(
        model=model,
        messages = chat_history,
        max_tokens= response_length,
        temperature = creativity,
        stop = [stop_phrase]
    )

    ai_res = res.choices[0].message.content
    print(f"AI : {ai_res}")
    chat_history.append({"role":"system", "content":ai_res})

# 대화 시작
chat(creativity, response_length, stop_phrase)
```
