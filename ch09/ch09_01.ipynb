### 지도 위치 정보 표시

```python
import folium
import pandas as pd

# 데이터 파일 업로드
from google.colab import files
uploaded = files.upload()

# 데이터 읽기
data = pd.read_excel('data.xls')

# Folium 지도 생성
m = folium.Map(location=[data['시설위도'].mean(), data['시설경도'].mean()], zoom_start=10)

# 위치 정보를 지도에 표시
for i in range(len(data)):
    folium.Marker([data.iloc[i]['시설위도'], data.iloc[i]['시설경도']],
                  popup=data.iloc[i]['시설명']).add_to(m)

# 지도 표시
m
```
